```python
#isolating data from 2017
salary_2017 = salary.loc[salary['Season End']==2017]
stats_2017 = seasonStats.loc[seasonStats['Year']==2017]
#Removing variables we dont want in our model
stats_2017 = stats_2017.drop(columns=['Unnamed: 0','FTr','ORB%','DRB%','TRB%','AST%','STL%','BLK%','blanl','OWS','DWS','WS','blank2','OBPM','DBPM','BPM','FGA','3P','2P','2PA','eFG%','FT','FTA','ORB', 'DRB'])
#replacing team abriviations of franchises that have changed cities/ names
#df['Team'].replace(['CHA','NJN','NOH'],['CHO','BRK','NOP'])
#df = pd.DataFrame(salary_2017, columns = ["Team",])
#renaming 'Player Names' in salary_2017 to 'Player'
salary_2017.rename(columns={'Player Name': 'Player'})
salary_2017.drop(columns=['Register Value','Season Start','Season End','Team','Full Team Name'])
```

```python
def truncate(n, decimals=1):
    multiplier = 10 ** decimals
    return int(n * multiplier) / multiplier

#Adding some Per Game stats
stats_2017['PPG'] = stats_2017['PTS']/stats_2017['G']
stats_2017['MPG'] = stats_2017['MP']/stats_2017['G']
stats_2017['SPG'] = stats_2017['STL']/stats_2017['G']
stats_2017['RPG'] = stats_2017['TRB']/stats_2017['G']
stats_2017['APG'] = stats_2017['AST']/stats_2017['G']
stats_2017['TOPG'] = stats_2017['TOV']/stats_2017['G']
stats_2017['BPG'] = stats_2017['BLK']/stats_2017['G']
stats_2017['PFPG'] = stats_2017['PF']/stats_2017['G']

#rounding
round(stats_2017['PPG']) 
round(stats_2017['MPG'])
round(stats_2017['SPG'])
round(stats_2017['RPG'] )
round(stats_2017['APG'])
round(stats_2017['TOPG'])
round(stats_2017['BPG'] )
round(stats_2017['PFPG'])
```
